**Ejemplo Calculadora Sencilla**

```python
print("Menú")
print()
print("1) Suma:+")
print("2) Resta:-")
print("3) Multiplicación:*")
print("4) División:/")
print()
opcion=input("Ingrese la opción deseada: ")
numero1=float(input("Ingrese el primer número: "))
numero2=float(input("Ingrese el segundo número: "))
print()
if opcion=="1" or opcion=="suma" or opcion=="+":
  print("Se suman los dos valores ingresados")
  print("La suma es:",numero1+numero2)
if opcion=="2" or opcion=="resta" or opcion=="-":
  print("Se restan los dos valores ingresados")
  print("La resta es:",numero1-numero2)
if opcion=="3" or opcion=="multiplicación" or opcion=="*":
  print("Se multiplican los dos valores ingresados")
  print("La multiplicación es:",numero1*numero2)
if opcion=="4" or opcion=="división" or opcion=="/":
  print("Se dividen los dos valores ingresados")
  if numero2!=0:
    print("La división es:",numero1/numero2)
  else:
    print("No se puede dividir entre 0")
```
