**LISTAS**

```python
#Listas
#Con listas usamos los corchetes []
Nombres=["Daniel","Yesenia","Andres","Carlos"]
edades=[26,26,30,40]
varios=["cc",10164,"talla",30]
print(Nombres)
print(edades)
print(varios)
```

```python
#Modificar un valor de una lista
Nombres[1]="Luisa"
print(Nombres)

#Agregar otro valor en la lista en la posición que uno quiera
edades.insert(3,38)
print(edades)

#Agregar otro valor en la lista en la última posición
Nombres.append("Juan")
print(Nombres)

#Eliminar un elemento (es el primero que aparezca con ese valor)
Nombres.remove("Daniel")
print(Nombres)
edades.remove(26)
print(edades)
#Eliminar un elemento con la posición que uno quiera
edades.remove(edades[2])
print(edades)
#otra forma
edades.pop(2)
print(edades)
```

**TUPLAS**

```python
#Tuplas
#Con las tuplas usamos parentesís ()
#La diferencia de estas con las listas, es que los valores que contienen no se pueden cambiar o modificar
coordenada=(75.25,245.56)
print(coordenada)
#Mostrar un solo valor de la tupla
longitud=coordenada[0]
print("La longitud es: ",longitud,"°N")
print("La latitud es: ",coordenada[1],"°O")
```

```python
nombre1=input("Ingresa tu nombre: ")
print("Hola", nombre1)
#convierte el número tipo texto a tipo flotante
edad1=float(input("Ingresa tu edad: "))
print("Tu edad es: ",edad1,"años")
#para redondear
op=round(edad1/7,2)
print("op: ",op)
#otra forma es dandole formato a la variable
op=edad1/7
print(f"{op:.2f}")
```

```python
notas=[3.5,4,5,2,1,1.6,4.7]
prom=sum(notas)/len(notas)
#sum es para sumar los valores de lista y len se usa para saber el valor de la longitud de la lista (cant.)
print(f"El promedio es: {prom:.2f}")
```

**TABLAS**

```python
#Importar bibliotecas
#Son bibliotecas para ya usarlo solamente, es decir estas nos van a brindar los diseños de las tablas o gráficas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Por el momento solo se va usar la primera, no importa si se importan varias y no se usan

#Crear la base datos
#Construir las listas de valores
value_n_list=["Andrew","Phillip","John"]
print(value_n_list)

value_w_list=[50, 78, 91.5] #[kg]
value_h_list=[1.76, 1.87, 1.69] #[m]

#Vamos a ver el diccionario
#Usar un diccionario de la forma ({key:value})
#con el formato ("str":list)
data_dict={"name":value_n_list,
           "weight":value_w_list,
           "height":value_h_list}
print(data_dict)

#Convierte un diccionario en un marco de datos
df=pd.DataFrame(data_dict) #pandas accede a su clase y solo requiere que pases el diccionario como entrada
#solo se debe crear el diccionario y ya con eso se crea la tabla

print()
print(df)

#Luego para que salga la tabla presentada bonita hay que llamar la tabla
print()
df
```
