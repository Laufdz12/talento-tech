**CICLO FOR CON LISTA**

```python
estudiantes=[]
#definimos el tamaño de la lista
num_estudiantes=int(input("Ingrese el número de estudiantes: "))

#Lo siguiente es un ciclo for, el cual permite que uno vaya ingresando datos hasta donde uno le indique o desee
for j in range(num_estudiantes):
  print("Codigo",j)
  nombre=input("Ingrese el nombre: ")
  #Arego los nombres por orden de llegada
  estudiantes.append(nombre)

  print(estudiantes)
```

**CICLO WHILE**

```python
#Se genera una lista por defecto, vacía, que se puede ir llenando o modificando
Empleados=[]
#Agregar datos a una lista sin tamaño definido

#bandera
e=False
#Mientras una condición sea verdad el se va a seguir repitiendo, si es falsa ya no se ejecuta más eso es el ciclo While
#Ciclo while, se ejecuta mientras se cumpla la condición
while e==False:
  try:              #Si no ocurre un error el va a realizar
    #aqui va a ocurrir el error------------------
    tamaño=int(input("Ingrese el tamaño de la lista: "))
    #-----------------------
    for i in range(tamaño):
      print("Num",i)
      nm=input("Ingrese el nombre del empelado/a: ")

      Empleados.append(nm)
    print(Empleados)
    e=True

  except ValueError:
      print("Error:Por favor, digita únicamente valores numéricos enteros")
```

**DATA FRAME**

```python
#Como crear un dataframe
import pandas as pd  #Se importa la libreria pandas (pd: es como llamamos a pandas para usarlo)

df=pd.DataFrame()  #Se crea un dataframe vacio #Se invoca al constructor (df:es como se llama el data frame)

#agregar columnas

#1ra forma
df["Nombre"]=None  #El none es para decir que aun no quiero agregar los valores en la columna, que me aparezca vacía
#2da forma
df=df.assign(Edad=None)
print(df)

#Consignar datos en el dataframe
#Crear listas
nombres=["Luisa","Daniela","Jorge"]
edades=[25,23,35]

df["Nombre"]=nombres
df["Edad"]=edades

print()
print(df)
print()
df
```

**EJERCICIO TABLA**

```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Crear la base datos
#Construir las listas de valores
value_n_list=["Andrew","Phillip","John"]
print(value_n_list)

value_w_list=[50, 78, 91.5] #[kg]
value_h_list=[1.76, 1.87, 1.69] #[m]

#Vamos a ver el diccionario
#Usar un diccionario de la forma ({key:value})
#con el formato ("str":list)
data_dict={"name":value_n_list,
           "weight":value_w_list,
           "height":value_h_list}
print(data_dict)

#Convierte un diccionario en un marco de datos
df=pd.DataFrame(data_dict) #pandas accede a su clase y solo requiere que pases el diccionario como entrada
#solo se debe crear el diccionario y ya con eso se crea la tabla

print()
print(df)

#Luego para que salga la tabla presentada bonita hay que llamar la tabla
print()
df
```
