## Cargar Dataset

### Subtask:
Cargar un dataset de ejemplo de una biblioteca de Python (por ejemplo, `seaborn`) para realizar el análisis.


**Reasoning**:
First, I'll import the seaborn library, then load the 'iris' dataset into a DataFrame named 'df', and finally display the first 5 rows to confirm the data has been loaded correctly.



```python
import seaborn as sns

df = sns.load_dataset('iris')

df.head()
```

## Estadísticas Numéricas

### Subtask:
Calcular y mostrar las estadísticas descriptivas (conteo, media, desviación estándar, mínimo, 25%, 50%, 75%, máximo) para todas las columnas numéricas del dataset.


**Reasoning**:
To calculate and display the descriptive statistics for all numerical columns, I will use the `.describe()` method on the DataFrame `df`.



```python
df.describe()
```

## Estadísticas Categóricas

### Subtask:
Calcular y mostrar las estadísticas descriptivas (conteo, número de valores únicos, valor más frecuente, frecuencia del valor más frecuente) para todas las columnas categóricas del dataset.


**Reasoning**:
To calculate and display descriptive statistics for all categorical columns, I will use the `.describe()` method on the DataFrame `df` and specify `include='object'`.



```python
df.describe(include='O')
```

## Estadísticas Completas

### Subtask:
Calcular y mostrar un resumen completo de las estadísticas descriptivas para todas las columnas del dataset, abarcando tanto numéricas como categóricas.


**Reasoning**:
To get a complete summary of descriptive statistics for all columns (numerical and categorical), I will use the `.describe()` method with the `include='all'` argument on the DataFrame `df` and print the result.



```python
df.describe(include='all')
```
