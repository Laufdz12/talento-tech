```python
#leer un csv y pasarlo a un dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

#cargar el csv
df=pd.read_csv("IRIS (1).csv")
#estadisitca descriptiva
df.describe()
df
```

```python
df.describe()
```

```python
#Funciones
#Media
media_sepal_length=df["sepal_length"].mean()
media_sepal_width=df["sepal_width"].mean()
#Desvación estandar
desv_sepal_length=df["sepal_length"].std()
print(f'media_sepal_length: {media_sepal_length:.3f}')
print(f'desv estandar: {desv_sepal_length:.3f}')
```

```python
plt.hist(df["sepal_length"], bins=20, color='orange', edgecolor='black')
plt.title("Histograma de sepal_length")
plt.xlabel("Valores/Values")
plt.ylabel("Frecuencia/Frequency")
plt.grid(True)
plt.show()
```

```python
#discrminación por especie
setosa=df[df["species"]=="Iris-setosa"]
setosa.head()
versicolor=df[df["species"]=="Iris-versicolor"]
versicolor.head()
virginica=df[df["species"]=="Iris-virginica"]
virginica.head()
```

```python
#libreria para funciones de probabilidad
from scipy.stats import norm
```

```python
x=setosa["sepal_length"]
y=norm(setosa["sepal_length"].mean(),setosa["sepal_width"].std()).pdf(x)

#grafico de barras
fig, ax=plt.subplots()
ax.bar(x,y,label="gráfico barras")
ax.grid(True)
```
