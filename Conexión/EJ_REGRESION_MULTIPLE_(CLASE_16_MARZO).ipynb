## Métricas de Regresión

Las métricas de regresión son fundamentales para evaluar el rendimiento de un modelo predictivo. Aquí se detallan las más comunes, incluyendo ejemplos y una visualización del histograma de errores.

### 1. Suma de los Cuadrados de los Errores (SSE o Sr)

*   **Definición**: Mide la suma de las diferencias al cuadrado entre los valores observados ($y_i$) y los valores predichos por el modelo ($\hat{y}_i$). Es una medida de la variabilidad no explicada por el modelo.
*   **Fórmula**: $Sr = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
*   **Interpretación**: Un valor de Sr más bajo indica un mejor ajuste del modelo a los datos. Penaliza fuertemente los errores grandes debido al cuadrado. Su unidad es el cuadrado de la unidad de la variable dependiente.
*   **Ejemplo**: Si los valores reales son [10, 12, 15] y las predicciones son [8, 13, 14], los errores son [2, -1, 1]. Los errores al cuadrado son [4, 1, 1]. El Sr sería $4+1+1=6$.

### 2. Error Cuadrático Medio (MSE)

*   **Definición**: Es el promedio de los errores al cuadrado. Se calcula dividiendo la Suma de los Cuadrados de los Errores (Sr) por el número de observaciones (n).
*   **Fórmula**: $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \frac{Sr}{n}$
*   **Interpretación**: Similar a Sr, un MSE más bajo indica un mejor rendimiento del modelo. Es más fácil de comparar entre modelos con diferentes números de observaciones. Penaliza los errores grandes más que los errores pequeños.
*   **Ejemplo**: Si el Sr es 6 para n=3 observaciones, el MSE sería $6/3 = 2$.

### 3. Raíz del Error Cuadrático Medio (RMSE)

*   **Definición**: Es la raíz cuadrada del MSE. Es una de las métricas más utilizadas porque tiene las mismas unidades que la variable dependiente (y), lo que facilita su interpretación directa.
*   **Fórmula**: $RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$
*   **Interpretación**: Un valor de RMSE más bajo indica un mejor ajuste del modelo. Al estar en la misma unidad que la variable objetivo, es más intuitivo entender el 'error promedio' del modelo. Por ejemplo, si predices precios en dólares, un RMSE de $5 significa que el error típico de tu predicción es de $5.
*   **Ejemplo**: Si el MSE es 2, el RMSE sería $\sqrt{2} \approx 1.414$.

### 4. Error Absoluto Medio (MAE o SAE/n)

*   **Definición**: Mide el promedio de las diferencias absolutas entre los valores observados y los valores predichos. Se calcula como la Suma de los Errores Absolutos (SAE) dividida por el número de observaciones (n).
*   **Fórmula**: $MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| = \frac{SAE}{n}$
*   **Interpretación**: A diferencia de MSE y RMSE, MAE no penaliza tanto los errores grandes y es menos sensible a los valores atípicos. Es una medida más robusta del error promedio y está en las mismas unidades que la variable dependiente.
*   **Ejemplo**: Usando el ejemplo anterior de errores [2, -1, 1], los valores absolutos son [2, 1, 1]. El SAE sería $2+1+1=4$. Para n=3, el MAE sería $4/3 \approx 1.333$.

### 5. Coeficiente de Determinación ($R^2$)

*   **Definición**: Mide la proporción de la varianza en la variable dependiente que es predecible a partir de la(s) variable(s) independiente(s). Indica qué tan bien las variables independientes explican la variabilidad de la variable dependiente.
*   **Fórmula**: $R^2 = 1 - \frac{Sr}{St}$ donde $St = \sum_{i=1}^{n} (y_i - \bar{y})^2$ (Suma Total de Cuadrados, que mide la variabilidad total de y).
*   **Interpretación**: Varía de 0 a 1. Un $R^2$ de 0.80 significa que el **80% de la variabilidad** de la variable dependiente puede ser explicada por el modelo. Un valor más alto indica un mejor ajuste. Un $R^2$ negativo puede ocurrir en regresión no lineal o cuando la intercepción se fija a cero, y significa que el modelo es peor que un modelo que solo predice la media de `y`.

### 6. Error Porcentual Absoluto Medio (MAPE)

*   **Definición**: Mide el tamaño del error en términos porcentuales. Es útil cuando se desea expresar el error en relación con el tamaño de los valores reales.
*   **Fórmula**: $MAPE = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100\%$
*   **Interpretación**: Un MAPE del 10% significa que el error promedio de la predicción es del 10% del valor real. Es fácil de entender y comunicar. Sin embargo, no se puede usar si los valores reales ($y_i$) son cero o muy cercanos a cero, ya que causaría una división por cero o valores muy grandes.
*   **Ejemplo**: Si un valor real es 100 y se predice 90, el error absoluto porcentual es $|(100-90)/100| = 0.1 = 10\%$. Si tienes varios de estos, el MAPE es el promedio.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Re-crear las listas x_list y y_list del ejemplo de regresión lineal simple
# Estos valores fueron utilizados en un ejemplo anterior donde se calcularon m y b.
# (Asumimos m=0.849 y b=1.51 de cálculos previos en un ejemplo de regresión lineal simple)
x_list=[1,2,2,3,4,4,5,6]
y_list=[2,3,4,4,4,6,5,7]

# Re-crear el DataFrame para el ejemplo de regresión lineal simple
df_simple_reg = pd.DataFrame({'x': x_list, 'y': y_list})

# Coeficientes de la regresión lineal simple previamente calculados
m = 0.849
b = 1.51

# Calcular los valores predichos (y_pred)
df_simple_reg['y_pred'] = (m * df_simple_reg['x']) + b

# Calcular el error (residuos)
df_simple_reg['error'] = df_simple_reg['y'] - df_simple_reg['y_pred']

# Calcular el Error Porcentual Absoluto (APE) para cada observación
# Manejar casos donde y_i es cero para evitar divisiones por cero en MAPE
df_simple_reg['ape'] = np.where(df_simple_reg['y'] == 0, np.nan, np.abs((df_simple_reg['y'] - df_simple_reg['y_pred']) / df_simple_reg['y']))

# Calcular el MAPE (promedio de APE, ignorando NaN si hubo divisiones por cero)
mape = np.nanmean(df_simple_reg['ape']) * 100
print(f"El MAPE para este ejemplo es: {mape:.2f}%")

# Generar el histograma de errores
plt.figure(figsize=(10, 6))
sns.histplot(df_simple_reg['error'], kde=True, bins=len(x_list) // 2 if len(x_list) > 1 else 1, color='skyblue')
plt.title('Histograma de Errores (Residuos)')
plt.xlabel('Error (y - y_pred)')
plt.ylabel('Frecuencia')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Error Cero') # Línea de referencia en cero
plt.legend()
plt.show()
```

## Interpretación del Histograma de Errores:

Un histograma de errores (o residuos) es una herramienta visual crucial para entender la distribución de los errores de un modelo de regresión. Idealmente, los errores deben estar distribuidos normalmente alrededor de cero, sin un patrón claro.

*   **Forma de la Distribución**: Lo ideal es que los errores sigan una distribución aproximadamente normal (en forma de campana) centrada en cero. Esto sugiere que el modelo no tiene sesgos sistemáticos.
*   **Centro de la Distribución**: Si el pico del histograma está cerca de cero, significa que la mayoría de las predicciones del modelo son cercanas a los valores reales.
*   **Simetría**: Una distribución simétrica alrededor de cero es deseable. Si la distribución está sesgada hacia la izquierda o la derecha, puede indicar un sesgo en el modelo (subestimando o sobrestimando consistentemente).
*   **Valores Atípicos (Outliers)**: Barras en los extremos del histograma que están lejos del centro pueden indicar la presencia de valores atípicos que el modelo no predice bien.

Observando el histograma generado, podemos evaluar si nuestros errores se distribuyen de una manera que sugiere un buen rendimiento del modelo de regresión lineal simple. En este ejemplo, con pocos puntos de datos, la distribución puede no ser perfectamente suave, pero nos da una idea general de cómo se agrupan los errores.
