```python
#Vamos a crear una tabla
import pandas as pd

y_list=[78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4]
x1_list=[7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10]
x2_list=[26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68]
x3_list=[6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8]
x4_list=[60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]

#Diccionario
dicc={'y':y_list,'x1':x1_list,'x2':x2_list,'x3':x3_list,'x4':x4_list}

#Dataframe
df=pd.DataFrame(dicc)

df
```

#PASOS
1) Convierteme la tabla en una matriz y reemplaza la primera columna con puros unos
2) Sacarle la transpuesta a la matriz: Dame la matriz transpuesta
3) Sacame la columna de y de la tabla original como un vector columna
4) Multiplicame la transpuesta por la matriz original
5) Dame la inversa del resultado anterior
6) Multiplicame lo anterior por la transpuesta por el vector de y

```python
#Convertir la tabla en una matriz
import numpy as np

# Convertir el DataFrame a una matriz NumPy, excluyendo la columna 'y'
X = df[['x1', 'x2', 'x3', 'x4']].to_numpy()

# Crear una columna de unos con el mismo número de filas que X
ones_column = np.ones((X.shape[0], 1))

# Reemplazar la primera columna de la matriz con la columna de unos
# Concatenar la columna de unos con el resto de las características (X)
matrix = np.concatenate((ones_column, X), axis=1)

print("Matriz con la primera columna reemplazada por unos:")
print(matrix)
```

```python
# Sacar la transpuesta de la matriz
matrix_transpose = matrix.T

print("Matriz Transpuesta:")
print(matrix_transpose)
```

```python
# Sacar la columna 'y' de la tabla original como un vector columna
y_vector = df['y'].to_numpy().reshape(-1, 1)

print("Vector Columna 'y':")
print(y_vector)
```

```python
# Multiplicar la transpuesta por la matriz original
product_transpose_matrix = np.dot(matrix_transpose, matrix)

print("Resultado de la multiplicación (Transpuesta * Original):")
print(product_transpose_matrix)
```

```python
# Dame la inversa del resultado anterior
inverse_product_transpose_matrix = np.linalg.inv(product_transpose_matrix)

print("La inversa de la matriz (Transpuesta * Original) es:")
print(inverse_product_transpose_matrix)
```

```python
# Multiplicar lo anterior (la inversa) por la transpuesta por el vector de y
coefficients = np.dot(np.dot(inverse_product_transpose_matrix, matrix_transpose), y_vector)

print("Los coeficientes resultantes (beta values) son:")
print(coefficients)
```

```python
# Desempaquetar los coeficientes en variables individuales
beta0 = coefficients[0, 0] # Coeficiente para el término de intercepción (el 'uno' que agregamos)
beta1 = coefficients[1, 0] # Coeficiente para x1
beta2 = coefficients[2, 0] # Coeficiente para x2
beta3 = coefficients[3, 0] # Coeficiente para x3
beta4 = coefficients[4, 0] # Coeficiente para x4

print(f"Beta0 (Intercepto): {beta0:.4f}")
print(f"Beta1 (Coeficiente x1): {beta1:.4f}")
print(f"Beta2 (Coeficiente x2): {beta2:.4f}")
print(f"Beta3 (Coeficiente x3): {beta3:.4f}")
print(f"Beta4 (Coeficiente x4): {beta4:.4f}")
```

```python
df
```

```python
#Agregar otra columna de y_pred=x*betas

df['y_pred']=round(beta0+df['x1']*beta1+df['x2']*beta2+df['x3']*beta3+df['x4']*beta4,3)

df
```

```python
#Agregar la columna de error
df['error']=round(df['y']-df['y_pred'],3)
df
```

```python
#Agregar columna error al cuadrado
df['error_cuadrado']=round(df['error']**2,3)
df
```

```python
#Sumar la columna de los errores
error_cuadrado_sum=df['error_cuadrado'].sum()
print('El total de los errores al cuadrado: ',error_cuadrado_sum)
```

```python
#Sacar el promedio de lo anterior
promedio_error_cuadrado=round(error_cuadrado_sum/len(df),3)
print('El promedio de los errores al cuadrado: ',promedio_error_cuadrado)
```

```python
#Raiz cuadrada del promedio
RMSE=round(promedio_error_cuadrado**0.5,3)
print('La raiz cuadrada del promedio de los errores al cuadrado: ',RMSE)
```

```python
#Le sacamos valor absoluto a la columna de errores
df['error_absoluto']=abs(df['error'])
df
```

```python
#Sumamos los valores absolutos
error_absoluto_sum=df['error_absoluto'].sum()
print('La suma de los errores absolutos: ',error_absoluto_sum)
```

```python
#Le sacamos el promedio a los valores absolutos MAE
MAE=round(error_absoluto_sum/len(df),3)
print('MAE es: ',MAE)
```

```python
# Calcular el Error Porcentual Absoluto Medio (MAPE)
# Primero, calculamos el error porcentual absoluto para cada observación
# Aseguramos que no haya división por cero en caso de que 'y' contenga valores cero

df['percentage_error'] = abs((df['y'] - df['y_pred']) / df['y']) * 100

# Ahora, calculamos el MAPE como el promedio de estos errores porcentuales absolutos
MAPE = round(df['percentage_error'].mean(), 3)

print(f"El Error Porcentual Absoluto Medio (MAPE) es: {MAPE}%")
```
